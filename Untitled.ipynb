```python
import pandas as pd
from sklearn.datasets import load_digits



#predict givin number curroctly predicting or not ?????
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
#plot images
import matplotlib.pyplot as plt
plt.gray()
#this will change image into gray
plt.matshow(digits.images[0])
```

```python
#plot 5 images
for i in range(5):
    plt.matshow(digits.images[i])
#printed numbers    
```

```python
#create data frame 
df=pd.DataFrame(digits.data)
df
```

```python
#add targets 
df['target']=digits.target
```

```python
df
```

```python
x=df.drop('target',axis='columns')
```

```python
x
```

```python
y=df.target
```

```python
y
```

```python
#split the data 

from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#now import random 

from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
#see here  n_estimators ---10 ------10 random trees
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)

#x test is predicted 
#y test is actual
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
model.score(x_test,y_test)
#50 random decision tres givin

#if random samplers incresing and accuracy is incresing
```

```python
#now will see confusion matrix
y_predicted=model.predict(x_test)
#we assigned our prediction to y_predicted
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
#will plot the graph
import seaborn as sns
# sns.heatmap(y_test,y_predicted)
sns.heatmap(cm,annot=True)
import matplotlib.pyplot as plt
plt.xlabel("predicted")
plt.ylabel("actual")
```

```python
#see there 35 is predicted well 
#then see 1 actual is 4 but my prediction is 1 
#see last 2 ctual is 4 but my prediction is 9 
#this is accuracy is differing
```
